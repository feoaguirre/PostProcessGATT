```python
def loadlog(name=None):
    print("load logs")
    
    # Check if the file name was provided
    if name is None:
        filename = 'log.txt'
    else:
        filename = name
    
    # Load the data from the file, skipping the header (assuming the header occupies 1 line)
    rawdata = np.loadtxt(filename, skiprows=1)
    
    # Separate the columns as specified in the MATLAB code
    log = {
        'saveNumber': rawdata[:, 0],
        'iter': rawdata[:, 1],
        't': rawdata[:, 2],
        'dt': rawdata[:, 3],
        'cfl': rawdata[:, 4],
        'res': rawdata[:, 5:10],  # Columns 6 to 10
        'U': rawdata[:, 10::5],   # Every 5 columns, starting from column 11
        'V': rawdata[:, 11::5],   # Every 5 columns, starting from column 12
        'W': rawdata[:, 12::5],   # Every 5 columns, starting from column 13
        'R': rawdata[:, 13::5],   # Every 5 columns, starting from column 14
        'E': rawdata[:, 14::5]    # Every 5 columns, starting from column 15
    }
    
    return log


log = loadlog()
```

```python
import numpy as np

def calculate_fft_1d(log):
    """
    Calculates the 1D FFTs along time for all columns of log['U'], log['W'], and log['R'],
    removing the mean of each column before computing the FFT.
    
    Parameters:
        log (dict): Dictionary containing the data loaded by loadlog().
    
    Returns:
        dict: A dictionary with the computed 1D FFTs for U, W, and R.
    """
    print("Calculating 1D FFTs...")
    
    # Extract the time vector
    t = log['t']
    
    # Solicitar ao usuário os valores de t_inicial e t_final
    try:
        t_inicial = float(input("\nEnter the initial time (t_inicial): "))
        t_final = float(input("Enter the final time (t_final): "))
        
        if t_inicial >= t_final or t_inicial < t[0] or t_final > t[-1]:
            raise ValueError("Invalid time range. Ensure t_inicial < t_final and within the time vector.")
        
        # Filtrar os índices correspondentes ao intervalo [t_inicial, t_final]
        idx_range = np.where((t >= t_inicial) & (t <= t_final))[0]
        t_filtered = t[idx_range]
        dt = np.mean(np.diff(t_filtered))  # Average time interval between samples
        
        # Frequencies corresponding to the FFT
        freq = np.fft.fftfreq(len(t_filtered), d=dt)
        freq_shifted = np.fft.fftshift(freq)  # Center the frequencies
        
        # Dictionary to store the FFT results
        fft_results = {
            'freq_shifted': freq_shifted,
            'U': [],
            'W': [],
            'R': []
        }
        
        # Helper function to compute the 1D FFT of a column
        def compute_fft(column):
            column_mean = np.mean(column)  # Compute the mean of the column
            column_detrended = column - column_mean  # Remove the mean
            fft_result = np.fft.fft(column_detrended)  # Compute the FFT
            fft_shifted = np.fft.fftshift(fft_result)  # Center the spectrum
            return fft_shifted
        
        # Process each variable (U, W, R)
        for var_name in ['U', 'W', 'R']:
            data_matrix = log[var_name][:, idx_range]  # Filter data for the selected time range
            fft_list = []  # List to store the FFTs of the columns
            
            # Iterate over the columns of the matrix
            for col_idx in range(data_matrix.shape[1]):
                column = data_matrix[:, col_idx]  # Extract the column
                fft_col = compute_fft(column)  # Compute the FFT of the column
                fft_list.append(fft_col)
            
            # Store the FFTs in the results dictionary
            fft_results[var_name] = np.array(fft_list).T  # Transpose to maintain consistency with dimensions
        
        print("1D FFTs calculated successfully for the selected time range.")
        return fft_results
    
    except ValueError as e:
        print(f"Error: {e}. Please enter valid values for t_inicial and t_final.")
        return None

# Calculate the 1D FFTs
fft_results = calculate_fft_1d(log)

if fft_results is not None:
    # Example of accessing the results
    freq_shifted = fft_results['freq_shifted']
    fft_U = fft_results['U']
    fft_W = fft_results['W']
    fft_R = fft_results['R']
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to find indices corresponding to a specific X, Y, or Z value
def find_indices_by_coordinate(probes, coordinate, value):
    """
    Finds the indices of rows in 'probes' where the specified coordinate matches the desired value.
    
    Parameters:
        probes: Coordinate matrix [Nx3].
        coordinate: Coordinate to filter ('X', 'Y', or 'Z').
        value: Desired coordinate value.
    
    Returns:
        list: List of matching indices.
    """
    coord_idx = {'X': 0, 'Y': 1, 'Z': 2}[coordinate]
    return [i for i, probe in enumerate(probes) if np.isclose(probe[coord_idx], value)]

# Function to find the column index corresponding to a specific (X, Y, Z) position
def find_column_index(X, Y, Z, probes):
    """
    Finds the column index corresponding to the position (X, Y, Z).
    
    Parameters:
        X, Y, Z: Desired coordinates.
        probes: Coordinate matrix [Nx3].
    
    Returns:
        int: Column index or None if not found.
    """
    for idx, probe in enumerate(probes):
        if np.isclose(probe[0], X) and np.isclose(probe[1], Y) and np.isclose(probe[2], Z):
            return idx
    return None

# Show available X values
print("Available X values:")
X_available = sorted(set(probes[:, 0]))
for i, X_val in enumerate(X_available):
    print(f"{i + 1}: X={X_val:.2f}")

# Request user input to choose an X value
try:
    choice_X = int(input("\nEnter the number corresponding to the desired X value: ")) - 1
    if choice_X < 0 or choice_X >= len(X_available):
        raise ValueError("Invalid X choice.")
    
    # Select the chosen X value
    X_chosen = X_available[choice_X]

    # Find the Y values corresponding to the chosen X
    indices_X = find_indices_by_coordinate(probes, 'X', X_chosen)
    Y_available = sorted(set(probes[indices_X, 1]))
    print(f"\nAvailable Y values for X={X_chosen:.2f}:")
    for i, Y_val in enumerate(Y_available):
        print(f"{i + 1}: Y={Y_val:.2f}")

    # Request user input to choose a Y value
    choice_Y = int(input("\nEnter the number corresponding to the desired Y value: ")) - 1
    if choice_Y < 0 or choice_Y >= len(Y_available):
        raise ValueError("Invalid Y choice.")
    
    # Select the chosen Y value
    Y_chosen = Y_available[choice_Y]

    # Find the index corresponding to the position (X, Y, Z=0)
    Z_chosen = 0.0
    col_idx = find_column_index(X_chosen, Y_chosen, Z_chosen, probes)
    if col_idx is None:
        raise ValueError("Position (X, Y, Z=0) not found in the data.")
    
    print(f"\nChosen position: X={X_chosen:.2f}, Y={Y_chosen:.2f}, Z={Z_chosen:.2f}")
    
    # Extract the FFTs for U, W, R at the chosen column
    fft_U_col = fft_results['U'][:, col_idx]
    fft_W_col = fft_results['W'][:, col_idx]
    fft_R_col = fft_results['R'][:, col_idx]
    freq_shifted = fft_results['freq_shifted']
    
    # Ensure the dimensions match by recalculating frequencies based on the FFT size
    n_fft = len(fft_U_col)  # Length of the FFT
    dt = log['dt'][0]  # Time step (assuming constant time step)
    freq = np.fft.fftfreq(n_fft, d=dt)  # Recalculate frequencies
    freq_shifted = np.fft.fftshift(freq)  # Center the frequencies
    
    # Filter only positive frequencies
    positive_freq_indices = freq_shifted > 0
    freq_positive = freq_shifted[positive_freq_indices]
    fft_U_col_pos = fft_U_col[positive_freq_indices]
    fft_W_col_pos = fft_W_col[positive_freq_indices]
    fft_R_col_pos = fft_R_col[positive_freq_indices]
    
    # Plot the FFTs on the same graph with log-log scale
    plt.figure(figsize=(10, 6))
    plt.loglog(freq_positive, np.abs(fft_U_col_pos), label='FFT(U)', color='blue')
    plt.loglog(freq_positive, np.abs(fft_W_col_pos), label='FFT(W)', color='green')
    plt.loglog(freq_positive, np.abs(fft_R_col_pos), label='FFT(R)', color='red')
    
    plt.title(f"FFT for Position (X={X_chosen:.2f}, Y={Y_chosen:.2f}, Z={Z_chosen:.2f})")
    plt.xlabel("Frequency (ω > 0)")
    plt.ylabel("Amplitude (log scale)")
    plt.grid(which='both', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

except ValueError as e:
    print(f"Error: {e}. Please enter a valid value.")
```
